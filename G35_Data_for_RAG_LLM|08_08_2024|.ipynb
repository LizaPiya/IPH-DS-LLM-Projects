```python
import spacy
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Load spaCy English model
nlp = spacy.load("en_core_web_sm")
```

```python
#take some fraction from filtered_df
# Take a 20% sample
#small_df = filtered_df.sample(frac=0.2)

# Or take a fixed number of samples
small_df = filtered_df_cleaned.sample(n=50000)
```

```python
small_df.head(100)
```

# **Count Words in Each Note**

```python
# Function to count words in a note
def count_words(note):
    doc = nlp(note)
    return len([token for token in doc if not token.is_punct and not token.is_space])

# Apply the function to count words
small_df['word_count'] = small_df['chartnotes'].apply(count_words)
print(small_df[['chartnotes', 'word_count']].head())
```

# **Filter Notes with At Least 5 Words**

```python
# Filter notes with at least 5 words
filtered_df = small_df[small_df['word_count'] >= 5]
print(filtered_df[['chartnotes', 'word_count']].head())
```

```python
filtered_df.shape
```

# **Analyze Sentence Lengths**

```python
# Function to calculate sentence lengths
def analyze_sentence_length(note):
    doc = nlp(note)
    sentence_lengths = [len([token for token in sent if not token.is_punct and not token.is_space]) for sent in doc.sents]
    return sentence_lengths

# Apply the function to calculate sentence lengths
filtered_df['sentence_lengths'] = filtered_df['chartnotes'].apply(analyze_sentence_length)
print(filtered_df[['chartnotes', 'word_count', 'sentence_lengths']].head())
```

# **Calculate Additional Matrices**

```python
# Calculate additional metrics
filtered_df['num_sentences'] = filtered_df['sentence_lengths'].apply(len)
filtered_df['total_words'] = filtered_df['sentence_lengths'].apply(sum)
filtered_df['avg_sentence_length'] = filtered_df['total_words'] / filtered_df['num_sentences']

# Verify that the columns have been added correctly
print(filtered_df[['chartnotes', 'word_count', 'num_sentences', 'total_words', 'avg_sentence_length']].head())
```

#**Print notes with highest word counts**

```python
# Find notes with the highest word counts
top_n = 1000  # Number of top records you want to retrieve
top_notes = filtered_df.nlargest(top_n, 'total_words')

# Display the top notes with the highest word counts
print(top_notes[['chartnotes', 'word_count', 'num_sentences', 'total_words', 'avg_sentence_length']])
```
